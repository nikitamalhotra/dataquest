```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
titanic = pd.read_csv('Dataquest/DataVis/train.csv')
```

```python
columns_to_keep = ['Survived', 
                   'Pclass', 
                   'Sex', 
                   'Age', 
                   'SibSp', 
                   'Parch', 
                   'Fare', 
                   'Embarked']

titanic = titanic[columns_to_keep]
titanic.dropna(inplace=True)
```

```python
# Generate seaborn plot of the Age column.
# The smoothed line is called a kernel density estimation (KDE)

sns.distplot(titanic['Age'])
plt.show()
```

```python
# Generate just the KDE. Setting the shade parameter to true shades area under the curve.

sns.kdeplot(titanic['Age'], shade=True)
plt.xlabel('Age')
plt.show()
```

```python
# We can use the seaborn.set_style() function to change the default 
# seaborn style sheet. Seaborn comes with a few style sheets:
# 
# darkgrid: Coordinate grid displayed, dark background color
# whitegrid: Coordinate grid displayed, white background color
# dark: Coordinate grid hidden, dark background color
# white: Coordinate grid hidden, white background color
# ticks: Coordinate grid hidden, white background color, ticks visible

# To remove the axis spines for the top and right axes, 
# we use the seaborn.despine() function

sns.set_style('white')
sns.kdeplot(titanic['Age'], shade=True)
plt.xlabel('Age')
sns.despine(left=True, bottom=True)
plt.show()
```

```python
# Condition on unique values of the "Survived" column.
g = sns.FacetGrid(titanic, col="Survived", size=6)
# For each subset of values, generate a kernel density plot of the "Age" columns.
g.map(sns.kdeplot, "Age", shade=True)
```

```python
# Use a FacetGrid instance to generate three plots on the same row:
# One for each unique value of Pclass.
# Each plot should be a kernel density plot of the "Age" column, with the area under the curve shaded.
# Each plot should have a height of 6 inches.
# Remove all of the spines using seaborn.despine().

g = sns.FacetGrid(titanic, col='Pclass', size=6)
g.map(sns.kdeplot, 'Age', shade=True)
sns.despine(left=True, bottom=True)
plt.show()
```

```python
# Using two parameters to generate FacetGrids

g = sns.FacetGrid(titanic, col="Pclass", row="Survived")
g.map(sns.kdeplot, "Age", shade=True)
sns.despine(left=True, bottom=True)
plt.show()
```

```python
g = sns.FacetGrid(titanic, 
                  col="Survived", 
                  row="Pclass", 
                  hue='Sex', size=3)
g.map(sns.kdeplot, "Age", shade=True).add_legend()
sns.despine(left=True, bottom=True)
plt.show()
```
